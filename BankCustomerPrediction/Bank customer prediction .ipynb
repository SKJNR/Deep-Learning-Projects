# Problem Statement

The goal here is to find weather the customer is going to leave the bank or not depending upon different features..

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
#loading data
df=pd.read_csv('BankCustomers.csv')
df.head()
```

```python
df.tail()
```

From here we can observe that 10000 rows has been contains.
   * I think first three features are not important to train model ,so i'm going to train model without these three models

```python
df=df.iloc[:,3:]
df.head()
```

# Data Preprocessing

From above we can observe that Gender and Geography are two different categorical features now i'm going to convert those features by applying label encoder.

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1:]
```

```python
#i'm just going to check data once again
X['Geography'].value_counts()
```

From above we can observe that there is only three countries .

```python
geography={'France':0,'Germany':1,'Spain':2}
X['Geography']=X['Geography'].map(geography)
```

```python
X['Gender'].value_counts()
```

```python
gender={'Male':0,'Female':1}
X['Gender']=X['Gender'].map(gender)
```

```python
#let's check data once again 
X.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing the Keras libraries and packages
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

# Creating Artificial Neural Network

```python
# Initialising the ANN
classifier = Sequential()
#FIRST HIDDEN +RELU LAYER
classifier.add(Dense(activation="relu", input_dim=10, units=6, kernel_initializer="uniform"))
#SECOND HIDDEN + RELU LAYER
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
#THIRD OUTPUT LAYER
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

# Making the predictions and evaluating the model

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```
